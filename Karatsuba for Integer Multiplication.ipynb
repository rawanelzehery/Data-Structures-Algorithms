```python
"""
Steps:
1-Divide the First number to 2 numbers by its middle of digit to A
and B
2- Divide the Second number to 2 numbers by its middle of digit to
C and D
3- Calculate the Following Terms:
C0 = A * C
C2 = B * D
C1 = (A + B) * (C + D) – C0 – C2
4- Result = C0 * 10n + C1 * 10n/2 + C2







"""
```

```python
def karatsuba(num1,num2):
    if num1 < 10 or num2 < 10 :
        return num1 * num2
    else:
        n = max(len(str(num1)),len(str(num2)))
        n2 = n // 2
        a = num1 // 10**n2  
        b = num1 % 10**n2
        c = num2 // 10**n2
        d = num2 % 10**n2
        c0 = karatsuba(a,c)
        c2 = karatsuba(b,d)
        c1 = karatsuba(a+b,c+d)-(c0+c2)
   
        return c0 * 10**(2*n2) + c1*10**n2 + c2
    
```

```python
n1 = 5432
n2 = 1678

print(n1*n2 == karatsuba(n1,n2))
```

```python
def karatsuba2(num1,num2):
    if num1 < 10 or num2 < 10 :
        return num1 * num2
    else:
        n = max(len(str(num1)),len(str(num2)))
        n2 = n // 2
        a , b = divmod(num1 , 10 **n2)
        c , d = divmod(num2 , 10 **n2)
        c0 = karatsuba2(a,c)
        c2 = karatsuba2(b,d)
        c1 = karatsuba2(a+b,c+d) - c0 - c2
        return c0 * 10 **(2 * n2)  + c1 * 10 **n2 + c2
        
        
    
```
