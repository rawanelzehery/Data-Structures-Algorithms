```python
#inplace merge sort with no return 
def merge(arr,start,middle,end):
    temp = []
    lst1 = start
    lst2 = middle+1
    while(lst1<=middle and lst2<=end):
        if arr[lst1] < arr[lst2]:
            temp.append(arr[lst1])
            lst1 = lst1 + 1
            
        else:
            temp.append(arr[lst2])
            lst2 = lst2 + 1
            
    while (lst1 <= middle):
            temp.append(arr[lst1])
            lst1 = lst1 + 1
            
    while (lst2 <=end):
            temp.append(arr[lst2])
            lst2 = lst2 + 1
            
    i = start
    while (i <= end ):
        arr[i] = temp[i - start]
        i = i + 1
        
    
            
            
        
    
    
```

```python
def merge_sort(arr,start_index,end_index):
    if end_index > start_index :
        middle_index = (start_index +end_index)//2
        merge_sort(arr,start_index,middle_index)
        merge_sort(arr,middle_index+1,end_index)
        merge(arr,start_index,middle_index,end_index)
        
      
    
        

    
```

```python
arr = [27,13,17,5,19]

start = 0
end = len(arr)-1

print("input_array : \n",arr)
merge_sort(arr,start,end)
print("sorted_array : \n",arr)
```
